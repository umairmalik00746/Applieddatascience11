# <strong>Italy Population</strong>: Comparision
Timeline: 1955 - 2020


```python
# Imporitng Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing Dataset
file = pd.read_excel("Italy_Population -1955-2020.xlsx")
# file = file[:5]
file
```

```python
# Converting Strings to Integers

total_world_population = file['Migrants (net)'].str.replace(",", "").astype(np.int64)
total_italy_Population = file['Population'].str.replace(",", "").astype(np.int64)
# yearly_chaging = file['Yearly % Change'].astype(np.int64)
year = file.loc[:,"Year"]

```
